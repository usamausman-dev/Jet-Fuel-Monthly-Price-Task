```python
import pandas as pd
file = "data.csv"
data_train = pd.read_csv(file)
```

```python
Months = data_train["Month"]
Price = data_train["Price"]
Change = data_train["Change"]
```

```python
import matplotlib.pyplot as plt
```

# Month and Price

```python
plt.figure(figsize=(20, 10))

# plotting the points
plt.plot(Months, Price, color='blue', linestyle='solid', linewidth = 1,
		marker='o', markerfacecolor='red', markersize=5)

# setting x and y axis range
plt.ylim(1,350)
plt.xlim(1,300)


# naming the x axis
plt.xlabel('Months')
# naming the y axis
plt.ylabel('Price')

# giving a title to my graph
plt.title('Jet Fuel Monthly Price - Pakistan Rupee per Gallon')

# function to show the plot
plt.show()
```

# Month and Change

```python
plt.figure(figsize=(20, 10))


plt.plot(Months, Change, color='green', linestyle='solid', linewidth = 1,
		marker='o', markerfacecolor='yellow', markersize=5)

# setting x and y axis range
plt.ylim(1,350)
plt.xlim(1,300)


# naming the x axis
plt.xlabel('Months')
# naming the y axis
plt.ylabel('Price')

# giving a title to my graph
plt.title('Jet Fuel Monthly Price - Pakistan Rupee per Gallon')

# function to show the plot
plt.show()
```

```python
plt.figure(figsize=(50, 20))
# plotting a bar chart
plt.bar(Months, Price, width = 0.5, color = ['red', 'green','yellow','blue'])

# naming the x-axis
plt.xlabel('Months')
# naming the y-axis
plt.ylabel('Prices')
# plot title
plt.title('Jet Fuel Monthly Price - Pakistan Rupee per Gallon')

# function to show the plot
plt.show()
```

```python
plt.figure(figsize=(50, 20))

# plotting points as a scatter plot
plt.scatter(Months, Price, label= "stars", color= "red", marker= "*", s=50)

# x-axis label
plt.xlabel('x - axis')
# frequency label
plt.ylabel('y - axis')
# plot title
plt.title('Scatter plot!')
# showing legend
plt.legend()

# function to show the plot
plt.show()
```
